```python
import sys
import os

# Get the absolute path to the scripts directory
current_dir = os.getcwd()
scripts_dir = os.path.join(os.path.dirname(current_dir), 'scripts')
sys.path.append(scripts_dir)

# Import all functions from the script
from correlation_and_stock_movement import (
    load_stock_data,
    load_news_data,
    calculate_sentiment,
    prepare_combined_data,
    calculate_correlation,
    plot_correlation
)
```

```python
# Load data
stock_file = '../yfinance_data/AAPL_historical_data.csv'
news_file = '../raw_analyst_ratings.csv'
stock_symbol = 'APL'

stock_df = load_stock_data(stock_file)
news_df = load_news_data(news_file)

combined_df = prepare_combined_data(stock_df, news_df, stock_symbol)
correlation, p_value = calculate_correlation(combined_df)

print(f"Correlation between sentiment and daily returns for {stock_symbol}: {correlation:.3f}")
print(f"P-value: {p_value:.3f}")

plot_correlation(combined_df, stock_symbol, correlation, p_value)
```
